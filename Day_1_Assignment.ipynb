# **Day 1 Assignment (Lets Upgrade Data Visualisation)**

### Importing Modules and Packages

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

# **Question 1**

## Definig Data samples

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

## Ploting the samples in b vs a form.

```python
plt.title('Graph of a on x-axis and b on y-axis')
plt.xlabel('a (x-axis)')
plt.ylabel('b (y-axis)')
plt.plot(a,b, linestyle='-.')
```

# **Question 2**

## Definig Data samples

```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175] #sales of company2
```

## Plotting Samples

```python
plt.title('Sales Trend of Company 1 and 2 in 7 days')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.plot(days, sales_1, label='Company 1')
plt.plot(days, sales_2, label='Company 2')
plt.legend(loc='lower right')
```
